```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
```

```python
# Generate random data for demonstration
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2 * X + 1 + np.random.randn(100, 1) * 2

# Fit the model
reg = LinearRegression().fit(X, y)
y_pred = reg.predict(X)
```

```python
# Create a scatter plot of the data points
scatter = px.scatter(x=X.flatten(), y=y.flatten(), title='Linear Regression Simulation', width=900, height=900)
line_trace = go.Scatter(x=X.flatten(), y=y_pred.flatten(), mode='lines', name='Regression Line', line=dict(color='red'))
scatter.add_trace(line_trace)
scatter.show()
```
