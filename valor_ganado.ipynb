```python
import random
import numpy as np
```

```python
# Define constants
NUM_SIMULATIONS = 1000  # Number of simulations to run
FACTOR_RANGE = (-10, 10)  # Range of factor values
IDEAL_PROJECT_VALUE = 100  # Ideal project value

# Perform Monte Carlo simulation
valores_proy = []
for _ in range(NUM_SIMULATIONS):
    a = np.random.uniform(*FACTOR_RANGE)
    b = np.random.uniform(*FACTOR_RANGE)
    c = np.random.uniform(*FACTOR_RANGE)
    d = np.random.uniform(*FACTOR_RANGE)

    valor_proy_nuevo = IDEAL_PROJECT_VALUE + a + b + c + d
    valores_proy.append(valor_proy_nuevo)

# Analyze and visualize results
mean_value = np.mean(valores_proy)
std_deviation = np.std(valores_proy)

print(f"Mean Project Value: {mean_value}")
print(f"Standard Deviation: {std_deviation}")

# You can also create a histogram to visualize the distribution
import matplotlib.pyplot as plt

plt.hist(valores_proy, bins=50, density=True, alpha=0.6, color='b')
plt.xlabel('Project Value')
plt.ylabel('Density')
plt.title('Monte Carlo Simulation Results')
plt.grid(True)
plt.show()
```
